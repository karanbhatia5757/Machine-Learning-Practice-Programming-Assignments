```python
def compute_cost(X_train,y_train,beta,m):
    #Write your code in place of None. Cost can be calculated using a single line of code
    cost=np.sum((np.dot(X_train,beta)-y_train) ** 2)/(2*m)
    
    return cost
```

```python
beta_testcase=np.zeros((8,1))
cost_verify= compute_cost(X_train,y_train,beta_testcase,m)
print(cost_verify)
```

```python
def gradient_descent(X_train,y_train,learning_rate,beta,m,num_iters):
    # In place of None, write the updated value of beta0 in temp0 and of beta1 in temp1
    for i in range(num_iters):
        # derivative vector is given by : X_train.Transpose *  (( X_train * betavector)- y ) 
        temp =   beta - (learning_rate/m) * np.dot(X_train.transpose(), (np.dot(X_train,beta)-y_train))
        beta = temp
       
        
        if(i%100==0):
            # In place of None, call the cost you just coded above
            cost= compute_cost(X_train,y_train,beta,m)
            print("Cost")
            print(cost)
             
            
            
    return beta      
```

```python
def stochastic_gradient_descent(X_train,y_train,learning_rate,beta,m,num_iters):

    for j in range(num_iters):
    
        for i in range(0,m):
        # Write updated value of beta vector in place of None
          
            temp = beta - (learning_rate)*(X_train[i,:].reshape((8,1)) * (np.dot(X_train[i,:] , beta) - y_train[i]))
            beta = temp
            
            
        
        if(j%100==0):
            
            cost= compute_cost(X_train,y_train,beta,m)
            print("Cost")
            print(cost)
            
    return beta
```

```python
def multiple_linear_reg_model_gda(X_train,y_train,m,learning_rate,num_iters):
    #initialize the values of parameter vector beta. It should be a column vector of zeros of dimension(m,1)
    beta= np.zeros((X_train.shape[1],1))
    
    #calculate the initial cost by calling the function you coded above.
    initial_cost=compute_cost(X_train,y_train,beta,m)
    print("Initial Cost")
    print(initial_cost)
    
    #Calculate the optimized value of gradients by calling the gradient_descent function coded above
    
    beta= gradient_descent(X_train,y_train,learning_rate,beta,m,num_iters)
    
    #Calculate the cost with the optimized value of beta0 and beta1 by calling the cost function.
    
    final_cost=compute_cost(X_train,y_train,beta,m)
    print("Final Cost")
    print(final_cost)
    return beta
```
